# Figure 2 Decay as a function of N

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../outputs/fig2a_2000samples.csv')
df
```

```python
fig, ax = plt.subplots(figsize=(12,12))

sns.lineplot(ax=ax, data=df, x='n_tot', y='error_angle_mean', hue='k', ci='error_angle_std', lw=3)
# plt.xscale("log")

plt.xlabel("Total number of samples ($n$)", fontsize=20)
plt.ylabel("Mean error (degrees)", fontsize=20)
plt.yticks(fontsize=18)
plt.xticks([0, 50, 100, 150, 200, 250, 300], fontsize=18)
plt.xlim(0, 300)
plt.ylim(0, 10)
plt.legend(title="Samples per site", loc="upper right", fontsize=18, title_fontsize=18)
ax.spines[['right', 'top']].set_visible(False)

axin = ax.inset_axes([0.30, 0.55, 0.42, 0.42])
sns.lineplot(ax=axin, data=df, x='n_tot', y='error_angle_mean', hue='k', ci='error_angle_std', lw=2, legend=False)
axin.set_xscale("log")
axin.set_yscale("log")
axin.set_xlabel("Total number of samples ($n$)", fontsize=16)
axin.set_ylabel("Mean error (degrees)", fontsize=16)
axin.spines[['right', 'top']].set_visible(False)

plt.savefig("Figure2a.pdf", format="pdf", bbox_inches='tight')
plt.savefig("Figure2a.png", format="png", bbox_inches='tight')
```
