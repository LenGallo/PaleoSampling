# Figure 3: Intersections

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.interpolate import InterpolatedUnivariateSpline

import smpsite as smp

%matplotlib inline
```

```python
df = pd.read_csv('../outputs/fig3_1000sim_Lean_100_morethanonestrategy.csv')

df_ = pd.read_csv('../outputs/fig3_1000sim_Lean_100_one_site_vandamme.csv')

df = pd.concat([df,df_])
```

```python
df_ = df.copy()

fig, ax = plt.subplots(figsize=(8,6))
# create grouped boxplot 
sns.boxplot(data = df_,
            x = 'samples_per_sites',
            y = 'error_angle',
            hue = 'outlier_rate',ax=ax)
ax.set_title("$n_{tot}$ ~ 100 samples")
```
